```python
#importing the libraries

import pandas as pd
import numpy as np
```

```python
#to read the csv file
df = pd.read_csv('raw_data.csv')
```

```python
# Task 1: Add information at index 9
new_row = pd.Series([210, 'Jim Carry', '1/1/1970', 'jim.carry@gmail.com', 'Male', 'Amazon', 'USA', 12349.556, 23652],
                    index=df.columns)
df = df.append(new_row, ignore_index=True)

# Task 2: Add missing emails at index 3 and 6
df.at[3, 'Email'] = 'graeme.croisier@gmail.com'
df.at[6, 'Email'] = 'modesta.batterson@gmail.com'

# Task 3: Replace invalid country names and fill NaNs in the 'Country' column
valid_countries = {'India', 'USA', 'Canada', 'Russia'}
df['Country'] = df['Country'].apply(lambda x: x if x in valid_countries else np.nan)
df['Country'].fillna('Unknown', inplace=True)

# Display the updated DataFrame
print(df)
```
